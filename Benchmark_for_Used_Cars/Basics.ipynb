## Exploring Odometer and Price

```python
autos.head()
```

We will be exploring price and odometer, but as seen in the table above they are numeric values with symbol or letters. We will remove those characters and turn them into integers along with the name "odometer" to "odometer_km".

```python
autos["price"] = (autos["price"]
                  .str.replace("$", "")
                  .str.replace(",", "")
                  .astype(int))

autos["price"].head()
```

```python
autos["odometer"] = (autos["odometer"]
                     .str.replace("km", "")
                     .str.replace(",", "")
                     .astype(int))

autos.rename({"odometer": "odometer_km"}, axis=1, inplace=True)
autos["odometer_km"].head()
```

```python
autos["odometer_km"].value_counts()
```

```python
print(autos["price"].unique().shape)
print("\n")
print(autos["price"].describe())
print("\n")
print(autos["price"].value_counts().head(25))
```

```python
autos["price"].value_counts().sort_index(ascending=False).head(25)
```

```python
autos["price"].value_counts().sort_index(ascending=True).head(25)
```

Since we are observing eBay, it is plausible for bidding to start at \\$1. We will remove all listings that is there for \\$0.

Through observation in the top 25 of prices the biggest jump is between \\$350,000 and \\$999,990. Since this is not as realistic we will remove all listings above \\$350,000.

```python
autos = autos[autos["price"].between(1, 350001)]
autos["price"].describe()
```

## Dealing with Incorrect Registration Year Data

In this section, we will analyze the column "registration_year" 

```python
autos["registration_year"].describe()
```

Above shows that some listing of cars had a registration year of 1000 and even 9999, years that cars have not been invented and years that have not passed.

We will look into the data further to select which listing should be taken out of the dataset.

```python
autos["registration_year"].value_counts(normalize=True, dropna=False).sort_index()
```

This column shows the year the car was registered. Having vehicles registered first before the listing is not possible and since all listings are in 2016, we will remove registered cars past 2016 from the data. The data also contains registered cars from previous centuries as early as year 1000, therefore I decided to take out years before 1900.

```python
autos = autos[autos["registration_year"].between(1900, 2016)]
```

```python
autos["registration_year"].value_counts(normalize=True, dropna=False).sort_index()
```

Here it shows that we have successfully removed registration_year before 1900 and after 2016. 

## Exploring Price by Brand

Here, we will focus more on the brands of the car.

```python
autos["brand"].value_counts(normalize=True)
```

There are many brands that has a small percentage of listing, therefore we will limit our brands to those that have more than 5% of the total listings.

```python
listing_brand = autos["brand"].value_counts(normalize=True)
top_brands = listing_brand[listing_brand > .05].index

print(top_brands)
```

We see that there is a total of 6 brands that has more than 5% of the total listing. Next, we will inspect their average price.

```python
brand_mean = {}

for brand in top_brands:
    company_brand = autos[autos["brand"] == brand]
    mean_price = company_brand["price"].mean()
    brand_mean[brand] = int(mean_price)
    
brand_mean
```

This shows that there is a clear difference in average pricing depending on the brand.
* Higher Average Price: Audi, Mercedes Benz, and BMW
* Middle Average Price: Volkswagen
* Lower Average Price: Ford, and Opel

## Exploring Mileage

Here, we will add the average mileage of the brand listed to compare it with the average price to get a better picture.

```python
bmp_series = pd.Series(brand_mean)
pd.DataFrame(bmp_series, columns=['mean_price'])
```

```python
brand_mean_mileage = {}

for brand in top_brands:
    company_brand = autos[autos["brand"] == brand]
    mean_mileage = company_brand["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_price = pd.Series(brand_mean).sort_values(ascending=False)
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_price, columns=["mean_price"])
brand_info
```

```python
brand_info["mean_mileage"] = mean_mileage
brand_info
```

The average mileage for all brands greater than 5% of the total listing seem to be similar. Most individuals tends to sell their vehicles between 120,000 to 135,000 miles but price of those vehicles depend on their brand name.
